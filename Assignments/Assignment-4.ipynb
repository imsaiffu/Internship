## 1. Scrape the details of most viewed videos on YouTube from Wikipedia. Url 
= https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos You need to find following details: 
A) Rank 
B) Name 
C) Artist 
D) Upload date 
E) Views

```python
page = requests.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')
page
```

```python
soup = BeautifulSoup(page.content)
soup
```

```python
table = soup.find("table", class_="wikitable")
```

```python
table
```

```python
ranks = []
names = []
artists = []
upload_dates = []
views = []
```

```python
for row in table.find_all("tr")[1:]:
    cells = row.find_all("td")
    
    if len(cells) >= 5:
        ranks.append(cells[4].text.strip())
        names.append(cells[0].text.strip())
        artists.append(cells[1].text.strip())
        upload_dates.append(cells[3].text.strip())
        views.append(cells[2].text.strip())

# Display the extracted data
for rank, name, artist, upload_date, view in zip(ranks, names, artists, upload_dates, views):
    print(f"Rank: {rank}, Name: {name}, Artist: {artist}, Upload date: {upload_date}, Views: {view}")
```

## 6. Scrape the details of Highest selling novels. 
A) Book name 
B) Author name 
C) Volumes sold 
D) Publisher 
E) Genre 
 Url - https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare

```python
page = requests.get("https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare")
page
```

```python
soup = BeautifulSoup(page.content)
soup
```

```python
table = soup.find("table", class_="in-article sortable")
table
```

```python
book_names = []
author_names = []
volumes_sold = []
publishers = []
genres = []
```

```python
for row in table.find_all("tr")[1:]:
    cells = row.find_all("td")
    
    if len(cells) >= 5:
        book_names.append(cells[1].text.strip())
        author_names.append(cells[2].text.strip())
        volumes_sold.append(cells[3].text.strip())
        publishers.append(cells[4].text.strip())
        genres.append(cells[5].text.strip())

# Display the extracted data
for book, author, volume, publisher, genre in zip(book_names, author_names, volumes_sold, publishers, genres):
    print(f"Book name: {book}, Author name: {author}, Volumes sold: {volume}, Publisher: {publisher}, Genre: {genre}")
```

```python
data = {
    "Book name": book_names,
    "Author name": author_names,
    "Volumes sold": volumes_sold,
    "Publisher": publishers,
    "Genre": genres
}
df = pd.DataFrame(data)
print(df)
```
